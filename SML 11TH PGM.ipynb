```python
from scipy import stats
import math

old_mean =200
sample_mean=190
sample_std=15
n=4
alpha=0.05

t_stat=(sample_mean - old_mean)/(sample_std/math.sqrt(n))

df=n-1

p_value=stats.t.cdf(t_stat,df=df)

print(f"T_statistic:{t_stat:.3f}")
print(f"P_value:{p_value:.5f}")

if p_value < alpha :
    print("Reject the null hypothesis: Thw new receipe has significantly fewer calories.")
else:
    print("Fail to reject the null  hypothesis: no significant difference in calories.")
    
```
