```python
# program to remove the vowels i.e give only consonants...
########  NOTE : Break can be only used inside while and for loop .. not for only conditional statements.
print("Enter x for Exit.")
string=input("Enter any string to remove All the vowels from it :")
if string == ' x':
    exit()
else:
    new_string = string
    print("Removing vowels from the given string...")
    vowels = {'a','e','i','o','u'}
    for char in string:
        if char in vowels :
            new_string=new_string.replace(char,"")
            
    print("The new_string After successfully removing all the vowels : ", str(new_string))
    
```

```python
# Same vowel removal with usauage if the function.
def remove_vowel(new_string,string):
    vowels = {'a','e','i','o','u'}
    for char in string.lower() :
        if char in vowels :
            new_string=new_string.replace(char,"")
    return new_string
print("Enter x for Exit.")
string=input("Enter any string to remove All the vowels from it :")
if string == ' x':
    exit()
else:
    new_string = string.lower()
    print("Removing vowels from the given string...")
    new_string=remove_vowel(new_string, string)
    print("The new_string After successfully removing all the vowels : ", new_string)
    
    
```

```python
###### LIST AND TUPLE Level 1######### 17 - 24
'''
List :
- collection of ordered values stored in a variable.
- It is mutable( changeable , can apply any operations)
- Lists are denoted using the [] brackets.
'''
#17. Multiplication of all elemts in the list.
my_list=[2,3,4,5]
multiplicand = 2
for i in range(len(my_list)):
    my_list[i]*=multiplicand

print(my_list)
```

```python
# Multiplication using the enumerate function
# ENUMERATE:   Is abuilt in function used in list to loop through and get both:
# Index and the element.

my_list=[2,3,4,5]
multiplicand = 2
for i , elem in enumerate (my_list):
    # enumerate==> [(0,2),(1,3),(2,4),(3,5)]
    
    my_list[i]= elem * multiplicand

print(my_list)
```

```python
22.# Remove matching elements .
my_list=[3,3,2,1]
element_to_remove= 3
if len(my_list)==0:
    print("Empty list")
elif my_list.count(element_to_remove)==0:
    print("Element not found")
else:
    for i in range(my_list.count(element_to_remove)): ### COUNT ==> No. of occurences.
        my_list.remove(element_to_remove)

print(my_list)
```
